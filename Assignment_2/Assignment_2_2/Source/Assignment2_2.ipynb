```python
import pandas
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pandas.read_csv('/content/drive/MyDrive/Colab Notebooks/Assignment 2/Assignment 2_2/Data_raw/diabetes.csv')
display(df)
```

A:

```python
seed_value = 15
sample_df = df.sample(n=25, random_state=seed_value)
display(sample_df)
```

```python
sample_df['Glucose'].describe()
```

```python
df['Glucose'].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Glucose'], color='blue', label='Population', kde=True, stat='density', alpha=0.5)
sns.histplot(sample_df['Glucose'], color='red', label='Sample', kde=True, stat='density', alpha=0.5)

plt.title('Distribution of Glucose: Population vs. Sample')
plt.xlabel('Glucose Level')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
df_mean_glucose = df['Glucose'].mean()
sample_df_mean_glucose = sample_df['Glucose'].mean()

df_max_glucose = df['Glucose'].max()
sample_df_max_glucose = sample_df['Glucose'].max()

data = {
    'Category': ['Population Mean Glucose', 'Sample Mean Glucose', 'Population Max Glucose', 'Sample Max Glucose'],
    'Value': [df_mean_glucose, sample_df_mean_glucose, df_max_glucose, sample_df_max_glucose]
}

comparison_glucose_df = pandas.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=comparison_glucose_df)
plt.title('Comparison of Glucose Mean and Max: Population vs. Sample')
plt.ylabel('Glucose Value')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

B:

```python
sample_98 = np.percentile(sample_df['BMI'], 98)
population_98 = np.percentile(df['BMI'], 98)

print("Sample 98th Percentile:", sample_98)
print("Population 98th Percentile:", population_98)
```

```python
data = {'Category': ['Sample 98th Percentile (BMI)', 'Population 98th Percentile (BMI)'],
        '98th Percentile Value': [sample_98, population_98]}
comparison_df = pandas.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='98th Percentile Value', data=comparison_df)
plt.title('Comparison of 98th Percentile BMI: Sample vs. Population')
plt.ylabel('98th Percentile BMI')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

C:

```python
sample_means = []

for i in range(500):
  df_bootstrap_sample = sample_df.sample(n=150, replace=True)
  sample_mean = df_bootstrap_sample['BloodPressure'].mean()
  sample_means.append(sample_mean)

plt.figure(figsize=(10, 6))
sns.histplot(sample_means, color='blue', kde=True, stat='density')
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
```

```python
average_sample = np.mean(sample_df['BloodPressure'])
standard_deviation_sample = np.std(sample_df['BloodPressure'])
std_rounded = round(standard_deviation_sample, 2)
sample_98_2 = np.percentile(sample_df['BloodPressure'], 98)
print(f"Average of sample: {average_sample}")
print(f"Standard deviation of sample: {std_rounded}")
print(f"98th Percentile of sample: {sample_98_2}")
```

```python
population_mean_bp = df['BloodPressure'].mean()
population_std_bp = df['BloodPressure'].std()
population_98_bp = np.percentile(df['BloodPressure'], 98)

avg_sample_means = np.mean(sample_means)
std_sample_means = np.std(sample_means)
sample_98_bp = np.percentile(sample_means, 98)


data = {'Category': ['Population Mean BloodPressure', 'Average of Sample Means (BloodPressure)'],
        'Mean Value': [population_mean_bp, avg_sample_means]}

comparison_df = pandas.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Mean Value', data=comparison_df)
plt.title('Comparison of Population Mean BloodPressure vs. Average of Sample Means')
plt.ylabel('Mean BloodPressure')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
population_std_bp = df['BloodPressure'].std()
std_sample_means = np.std(sample_means)
data_2 = {'Category': ['Population Standard Deviation BloodPressure', 'Standard Deviation of Sample Means (BloodPressure)'],
          'Standard Deviation': [population_std_bp, std_sample_means]}

comparison_df_2 = pandas.DataFrame(data_2)

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Standard Deviation', data=comparison_df_2)
plt.title('Comparison of Population Standard Deviation BloodPressure vs. Standard Deviation of Sample Means')
plt.ylabel('Standard Deviation BloodPressure')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
population_98_bp = np.percentile(df['BloodPressure'], 98)
sample_98_bp = np.percentile(sample_means, 98)

data_3 = {'Category': ['Population 98th Percentile BloodPressure', '98th Percentile of Sample Means (BloodPressure)'],
          '98th Percentile': [population_98_bp, sample_98_bp]}

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='98th Percentile', data=pandas.DataFrame(data_3))

plt.title('Comparison of Population 98th Percentile BloodPressure vs. 98th Percentile of Sample Means')
plt.ylabel('98th Percentile BloodPressure')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

For the Mean, this is normal for it to match closely to the population's. The standard deviation and 98th percentile are both affected due to the number of elements in each group. These are all due to the Central Limit Theorem.
