```python
import numpy
import pandas
import sklearn
```

```python
df = pandas.read_csv("/content/drive/MyDrive/Colab Notebooks/Assignment 2/Assignment 2_1/Raw_Data/train.csv")
df
```

```python
df_cleaned = df.dropna(subset=['New_Price'])
display(df_cleaned)
```

I dropped the NaN of the New_Price column due beacause of how few values there were in the column. With there only being 815 out of 5847, it wouldnt have been possible to fill each of the missing values.

```python
filtered_df = df_cleaned[(df_cleaned['Kilometers_Driven'] == 50000) & (df_cleaned['Fuel_Type'] == 'Electric')]
display(filtered_df)
```

```python
rows_with_nan = df_cleaned[df_cleaned.isnull().any(axis=1)]
display(rows_with_nan)
```

```python
df_cleaned = df_cleaned.drop(4313)
display(df_cleaned)
```

Row 4313 was dropped since no other row would relate to it, which made it so I couldnt impute it.

```python
df_cleaned['Mileage'] = df_cleaned['Mileage'].str.replace(' kmpl', '')
df_cleaned['Mileage'] = df_cleaned['Mileage'].str.replace(' km/kg', '')
df_cleaned['Engine'] = df_cleaned['Engine'].str.replace(' CC', '')
df_cleaned['Power'] = df_cleaned['Power'].str.replace(' bhp', '')
df_cleaned['New_Price'] = df_cleaned['New_Price'].str.replace(' Lakh', '')
df_cleaned['New_Price'] = df_cleaned['New_Price'].str.replace(' Cr', '')
df_cleaned
```

```python
df_cleaned = pandas.get_dummies(df_cleaned, columns=['Fuel_Type', 'Transmission'], dtype=int)
display(df_cleaned)
```

```python
df_cleaned['Car_Age'] = 2025 - df_cleaned['Year']
display(df_cleaned)
```

dplyr in Python:

1) Select:

```python
selected_df = df_cleaned[['Name', 'Year']]
display(selected_df)
```

2) Filter:

```python
selected_df = df_cleaned[(df_cleaned['Kilometers_Driven'] > 100000) & (df_cleaned['Price'] > 4) & (df_cleaned['Car_Age']  > 10)]
display(selected_df)
```

3) Rename:

```python
renamed_df = df_cleaned.rename(columns={'Name': 'Model_Name', 'Seats':'No. of Seats'})
display(renamed_df)
```

4) Mutate

```python
df_2 = df_cleaned.copy()
df_2['Price_Per_Kilometer'] = df_2['Price']/df_2['Kilometers_Driven']
display(df_2)
```

5) Arrange


```python
arranged_df = df_cleaned.sort_values(by='Year', ascending=True)
display(arranged_df)
```

6) Summarize with Group By:

```python
summarized_df = df_cleaned.groupby('Name')['Kilometers_Driven'].mean()
display(summarized_df)
```
