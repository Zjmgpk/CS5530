```python
import pandas as pd
import numpy as np
import category_encoders as ce
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assignment 1/Data_Raw/assignment_1_raw_data.csv')
df
```

```python
df = df.rename(columns={"Height ": "Height", "Weight ": "Weight"})
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
df['Height'] = df['Height'] * 0.0254
df['Weight'] = df['Weight'] * 0.42359237
print(df)
```

```python
df.columns = df.columns.str.strip()
df = df.rename(columns={"Height":"Height_m", "Weight":"Weight_kg"})
df['BMI'] = df['Weight_kg'] / (df['Height_m'] * 2)
df["BMI"] = df["BMI"].astype(float)
df["BMI"] = df["BMI"].round(2)
df
```

```python
bins = [0, 29, 45, 60, np.inf]
labels = ['<30', '30-45', '46-60', '60+']
df['Age Group'] = pd.cut(df['Age'].str.strip().astype(int), bins=bins, labels=labels, right=True)
table = pd.get_dummies(df['Age Group'], columns=['Age Group'], dtype=int)
table
```
